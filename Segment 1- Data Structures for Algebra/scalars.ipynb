## Scalars in TensorFlow 
#### Tensors created with a wrapper. 
#### . tf.variable
#### . tf.constant
#### . tf.placeholder
#### . tf.SparseTensor

#### Most widely-used is tf.variable. 
#### A with TF tensors, in PyTorch we can similarly perform operation, and we can easily convert to and from NumPy arrays.

```python
! pip install tensorflow
```

```python
import tensorflow as tf
```

```python
x_tf = tf.Variable(25, dtype=tf.int16) # dtype is optional
print(x_tf)
```

```python
x_tf.shape
```

```python
y_tf = tf.Variable(3, dtype=tf.int16)
```

```python
(x_tf + y_tf)
```

```python
tf_sum = tf.add(x_tf , y_tf)
print(tf_sum)
```

```python
tf_sum.numpy() # note that NumPy operation automatically convert tensors to NumPy arrays, and vice versa
```

```python
type(tf_sum.numpy())
```
