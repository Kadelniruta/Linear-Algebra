### Question: Consider the following examples:
##### . Sheriff has 180km/hr car
##### . Bank robber has 150km/hr car and five min head start
##### . How long does it take the sheriff to catch the robber?
##### . What distance will they have travelled at that point?
##### . (for simplicity, let's ignore acceleration, traffic etc)

### Solution: 

##### Bank Robber d = 2.5 km/min
##### Sheriff  d = 3 km/min

##### equation 1: d = 2.5t  
##### equation 2: d = 3 (t - 5)

##### or, 2.5t = 3 (t - 5)
##### or, 2.5t = 3t - 15
##### or, 2.5t - 3t = -15
##### or, -0.5t = -15
##### therefore, t = 30 , it means it  takes 30min long to sheriff to catch the robber.

##### Then, 
##### . d = 2.5 * 30 = 75 km
##### . d = 3 * (30 - 5) = 75 km
##### They have travelled at 75km. 



# Code

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,1000) # start, finish, n points #create point from starting to ending i.e linspace
```

### Distance travelled by robber: d = 2.5t

```python
d_r = 2.5 * t
```

### Distance travelled by sheriff: d = 3 (t-5)

```python
d_s = 3 * (t - 5)
```

```python
fig,ax = plt.subplots() # fig = entire figure , ax = represent the individual plot or axes with in the figure
plt.title("A bank Robber Caught")
plt.xlabel("time (in minutes)")
plt.ylabel("distance (in km)")
ax.set_xlim([0,40]) # lim[min, max] : define range of the x-axis
ax.set_ylim([0,100])
ax.plot(t,d_r, c="green") # Robber's distance in green
ax.plot(t,d_s, c ="red") # Sheriff's distance in red
plt.axvline(x = 30, color= "purple", linestyle = "--") # vertical line marking the time when sheriff catches the robber(30min).
__ = plt.axhline(y=75, color ="purple", linestyle = "--") # horizontal line marking the distance at which the robber is caught (75km).
plt.legend()
plt.grid()
plt.show()
```

# Exercise

#### Jill design solar panel as a hobby.
#### On April 1st jill's "Mark I" design begins generating power: 1KJ/day.
#### On May 1st, her "Mark II" design gegins generating 4KJ of power per day.

##### 1. What day is it when jill's mark II design has henerated as much total energy as the Mark I design?
##### 2. How much tota energy have both generated by that day?
##### 3. What would the solution to (.1) and (.2) be if mark II design generated 1KJ of power per day?

## Solution: 
## 1.
##### Mark I design starts on April 1 genrating 1KJ/day= total energy until t , E1 = t
##### Mark II starts on May 1, which is 30 days after April 1, generating 4KJ/day = total energy on day t = E2 = 4 (t - 30), valid only if t>=30.

##### Then, 
##### E1 = E2
##### t = 4 (t - 30)
##### 4t - 120 = t
##### 3t = 120
##### t = 40

##### So, on day 40 (10 days after May 1), Mark II's total energy equals Mark I's total energy.

## 2. 
##### Total Energy Generated: 
##### On day 40: 
#####    E1 = t = 40KJ
#####    E2 = 4 (t - 30) = 4 (40-30) = 4 * 10 = 40 KJ

##### Combined total energy:
#####    Total Energy = E1 + E2 = 40 + 40 = 80 KJ

## 3. 
##### No solution if Mark II and Mark I design generating same 1KJ of powr per day.

# Code

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,50,1000)
```

```python
E1 = t # mark I energy (1KJ/day)
```

```python
E2 = np.where(t>=30, 4 * ( t -30),0) # Energy genrated by mark II(starts genrating from day 30)
```

```python
fig,ax = plt.subplots()
plt.title("Energy generated by Mark I and Mark II Designs")
plt.xlabel("Days from April 1") 
plt.ylabel("Energy (KJ)")
ax.set_xlim([0,50])
ax.set_ylim([0,100])
ax.plot(t, E1 , c="green") # mark I energy curve
ax.plot(t, E2, c="red") # mark II energy curve
plt.axvline (x=40, color="purple", linestyle = "--") # equal time
__ = plt.axhline(y=40 , color = "purple", linestyle = "--") # equal energy
plt.grid()
plt.show()
```
